# Machine Learning

## Feature Vectors

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler

feature_assembler = VectorAssembler(inputCols=["open", "high", "low", "close"],
                                    outputCol="Independent Features")
```

## Model and Pipeline

```python
from pyspark.ml.regression import LinearRegression, RandomForestRegressor
```

## Random Forest Regressor

```python
# REGRESSOR
regressor = RandomForestRegressor(featuresCol='Independent Features', labelCol="adjclose")
```

```python
# PIPELINE
from pyspark.ml import Pipeline

pipeline = Pipeline(stages=[feature_assembler, regressor])

# Saving the pipeline
# pipeline.write().overwrite().save('pipeline_saved_model')

# Loading the pipeline
# pipelineModel = Pipeline.load('./pipeline_saved_model')


# SPLITTING THE DATA
train_data, test_data = appl.randomSplit([0.8, 0.2])

# TRAINING THE PIPELINE
model = pipeline.fit(train_data)

# PREDICTION
pred = model.transform(train_data)

# MODEL EVALUATION
from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(labelCol="adjclose")
RMSE = evaluator.evaluate(pred, {evaluator.metricName: 'rmse'})
R2 = evaluator.evaluate(pred, {evaluator.metricName: 'r2'})
MAE = evaluator.evaluate(pred, {evaluator.metricName: 'mae'})


RESULT = f"""
        RESULT
===========================================
RMSE   {RMSE}
MAE    {MAE}
R2     {R2}
"""

print(RESULT)
```
